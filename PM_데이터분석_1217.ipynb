## key로 받아오기

```python
api_key = api

# 요청 주소 = base_url + endpoint
base_url = 'https://api.themoviedb.org/3'
endpoint = '/movie/now_playing'
request_url = base_url + endpoint

# 쿼리 파라미터
params = {
    'api_key': api_key,
    'language': 'ko-KR'
}

# 헤더 : 데이터를 나타내는 데이터
# accept : 응답으로 받고 싶은 데이터를 명시
headers = {'accept' : 'application/json'}


# 요청 보내기
response = requests.get(request_url, params=params, headers=headers)
print(response)
# 200 = 정상, 외에는 오류
print(response.status_code)

pprint(response.json())
```

### 현재 상영 중인 영화 목록을 1 페이지부터 5 페이지까지 요청하고, 응답 데이터 리스트에 저장하고, 출력하기

```python
api_token = token

base_uel = 'https://api.themoviedb.org/3'
endpoint = f'/movie/now_playing'

all_page = []
for i in range(1, 6):
    request_url = base_url + endpoint

    params = {
            'language' : 'ko-KR',
            'page' : i
        }

    headers = {
        'Authorization': f'Bearer {api_token}',
    }

    response = requests.get(request_url, headers=headers, params = params)
    response_data = response.json()
    results = response_data['results']

    all_page.append(response_data)

# print(response)
# pprint(response.json())
# pprint(response.json()['results'])
# pprint(response.json()['results'][:3])
    pprint(results)
```

### 현재 상영 중인 영화 중 평균 평점(vote_average)이 가장 높은 영화 찾기
- 페이지부터 5페이지까지의 영화 중
해당 영화의 평균 평점과 영화 이름 출력

```python
api_token = token

base_uel = 'https://api.themoviedb.org/3'
endpoint = f'/movie/now_playing'

all_page = []
for i in range(1, 6):
    request_url = base_url + endpoint

    params = {
            'language' : 'ko-KR',
            'page' : i
        }

    headers = {
        'Authorization': f'Bearer {api_token}',
    }

    response = requests.get(request_url, headers=headers, params = params)
    response_data = response.json()

    all_page.append(response_data)

max_vote = 0
max_title = ''
for page in all_page:
    for movie in page['results']:
        if movie['vote_average'] > max_vote:
            max_vote = movie['vote_average']
            max_title = movie['title']

print(max_title, max_vote)
```

### 현재 상영 중인 영화 중 평균 평점(vote_average)이 가장 높은 영화의 리뷰 데이터 요청하고, 응답 데이터 출력하기

```python
import requests
from pprint import pprint

api_token = token
headers = {'Authorization': f'Bearer {api_token}'}

max_vote = 0
max_title = ''
max_movie_id = None

for i in range(1, 6):
    endpoint = f'/movie/now_playing'
    request_url = base_url + endpoint
    params = {'language': 'ko-KR', 'page': i}

    response = requests.get(request_url, headers=headers, params=params)
    response_data = response.json()

    for movie in response_data['results']:
        if movie['vote_average'] > max_vote:
            max_vote = movie['vote_average']
            max_title = movie['title']
            max_movie_id = movie['id']

endpoint = f'/movie/{max_movie_id}/reviews'
request_url = base_url + endpoint
response = requests.get(request_url, headers=headers)

print(response)
pprint(response.json())
```

### 가장 높은 평점의 영화 리뷰 출력

```python
import requests
from pprint import pprint

api_token = token

# 요청 주소 = base_url + endpoint
base_url = "https://api.themoviedb.org/3"
endpoint = "/movie/top_rated"
request_url = base_url + endpoint

# 쿼리 파라미터
params = {
    "language": "ko-KR",
}

# 인증 정보
headers = {
    "accept": "application/json",
    "Authorization": f"Bearer {api_token}",
}

response_data = requests.get(request_url, params=params, headers=headers)

response_python = response_data.json()

# 영화 정보(results) 추출
results = response_python["results"]

# 평가가 좋은 영화 중 첫 번째 영화
top_movie = results[0]

# 최고 평가 영화의 movie_id 추출
movie_id = top_movie["id"]

pprint(movie_id)

# 영화의 review를 요청하는 주소(URL)
base_url = "https://api.themoviedb.org/3"
endpoint = f"/movie/{movie_id}/reviews"
request_url = base_url + endpoint

# review 데이터 요청
response_reviews = requests.get(request_url, headers=headers)
pprint(response_reviews.json())
```
