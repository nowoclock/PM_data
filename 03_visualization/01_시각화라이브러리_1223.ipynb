### (1) Figure 와 Axes

`Figure`

- 전체 그래프의 컨테이너.
- 하나의 Figure는 여러개의 Axes를 가질 수 있다.
- 크기, 레이아웃, 제목을 설정할 수 있다.

`Axes`

- 실제 데이터가 시각화 되는 공간.
- 축(Axis), 레이블, 제목, 그리드를 설정할 수 있다.

```python
import matplotlib.pyplot as plt

# 그래프의 구성요소 (데이터)
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, label="y")
# 타이틀 설정
plt.title("Line Plot")
# 라벨 설정
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# 범례 표시
plt.legend()

# 그래프 출력
plt.show()
```

## [1] Scatter Plot (산점도)

### (1) **주요 특징**

- **데이터의 분포와 관계 파악**
    - 두 변수 간의 관계를 시각적으로 파악할 수 있다.
    - 예를 들어 변수 간에 강한 상관 관계가 있는지 아니면 관계가 없는지를 확인할 수 있다.
- **이상치 확인**
    - 데이터의 이상치나 특이한 패턴을 시각적으로 식별하기 쉽다.
- **데이터 밀도**
    - 많은 수의 데이터 포인트가 특정 영역에 모여 있으면 해당 영역에서 데이터의 밀도가 높다는 것을 의미한다.
- **색상 및 크기**
    - 산점도에서 점의 색상이나 크기를 조절하여 추가 변수의 정보를 표현할 수 있다.
- **적합한 데이터**
    - 두 개의 연속형 데이터
    - 두 개의 연속형 데이터를 구분하는 범주형 데이터

### 기본 사용

```python
# tips 데이터를 활용한 산점도
# 모듈 불러오기
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 셋 불러오기
tips_df = sns.load_dataset("tips")
# 정보 확인
tips_df.info()
```

```python
tips_df.head()
```

### 산점도

```python
# 2개의 연속형 데이터 기반

# total_bill 에 따른 tup을 산점도 시각화
sns.scatterplot(
    data=tips_df,
    x = "total_bill",
    y = "tip"
)
# 그래프 출력
plt.show()
```

### hue 매개변수

- 점을 구분하는 범주형 데이터

```python
sns.scatterplot(
    data=tips_df,
    x = "total_bill",
    y = "tip",
    hue = "size"
)

plt.show()
```
