## 범주형 데이터 타입

```python
import pandas as pd

df = pd.DataFrame(
    {
        "등급": [
            "하",
            "중",
            "상",
            "하",
            "상",
            "중",
            "하",
            "하하"
        ]
    }
)

df_categorical = pd.DataFrame(df["등급"])
df.info()
```

###category 타입으로 변환
- 순서    
     - 데이터 (범주) 리스트
     - 데이터 타입 변환할 열 지정
     - pd.Categorical()을 활용하여 변환
- 리스트 값 순서대로 범주형 데이터의 순서가 정해진다.

```python
# 하 < 중 < 상
# 1. 범주 리스트
orders = ["하", "중", "상"]
# 2. 변환 열 지정
column = df["등급"]
# 3. 변환
pd.Categorical(
    values=column, # 1. 변환할 열
    categories=orders, # 2. 리스트
    ordered=True, # 3. 크기비교 설정
)
# 하하는 지정하지 않았기 때문에 제외된다.
```

```python
# 등급 별로 순서를 부여함.
df.sort_values(by="등급")
```

```python
# 범주형으로 변환 class
# 1. 범주 리스트
orders = ['unacc', 'acc', 'good', 'vgood']

# 2. 변환 열 지정
column = df["class"]
# 3. 변환
# 새롭게 지정해줘야 저장된다.
df['class'] = pd.Categorical(
    values=column, # 1. 변환할 열
    categories=orders, # 2. 리스트
    ordered=True, # 3. 크기비교 설정
)
```

```python
# 범주형으로 변환 price
orders = ['low', 'med', 'high', 'vhigh']
column = df['price']
df['prince'] = pd.Categorical(
    values=column,
    categories=orders,
    ordered=True,
)
```
