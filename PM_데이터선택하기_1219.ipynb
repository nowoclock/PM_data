## 복수조건 필터링
- &, | 을 사용
- 각각을 ()로 묶어야 함
- **`Dataframe[(조건식1)&(조건식2)]`**
- **`Dataframe[(조건식1)|(조건식2)]`**

## 중복 데이터 삭제

### **중복 행 확인하기**

**`duplicated()`**

- 각 행이 중복인지 나타내는 Boolean Series 객체를 반환한다.
- 첫 번째 등장하는 행은 중복으로 식별되지 않으며, 그 이후에 동일한 행이 나타나면 중복으로 식별한다.
- **`keep`** 인자
    - 중복 행을 어떻게 식별할지 설정한다.
    - **`keep='first'`** (기본값) : 첫 번째 등장을 제외한 나머지 중복을 True로 식별한다.
    - **`keep='last'`** : 마지막 등장을 제외한 나머지 중복을 True로 식별한다.
    - **`keep=False`** : 모든 중복 행을 True로 식별한다.

```python
# 중복 데이터 중 첫 번째 데이터를 제외하고, True
df.duplicated(keep="first")
```

```python
# 중복 데이터 중 마지막 데이터를 제외하고, True
print(df.duplicated(keep="last"))
```

```python
# 중복 행을 찾아서 출력한다.
print(df.loc[df.duplicated()])
```

### **중복 행 삭제하기**

**`drop_duplicates()`**

- 데이터 프레임에서 중복 행을 삭제하고, 중복이 제거된 새로운 데이터 프레임을 반환한다.
- 이 메서드는 기본적으로 모든 열을 대상으로 중복을 확인한다.

**`subset`** 인자
-  해당 데이터가 중복인 경우 삭제

```python
# 모든 열이 동일한 중복 데이터 삭제
# 기존 데이터 프레임 삭제 x
df.drop_duplicates()

# keep인자
df.drop_duplicates(keep="last")

#subset 인자
df.drop_duplicates(subset=["나이", "직업"])
```
