# 문자열 처리

## 문자열 접근자 (str)
- 문자열 메서드를 활용하기 위한 도구

```python
df['이름'].str
```

### 대소문자 변환
- .str.upper() / .str.lower()
* 데이터 프레임을 직접 수정하지 않는다. *

```python
# 이메일 데이터를 통일
df['이메일'] = df['이메일'].str.lower()
df
```

### 특정 문자열 포함 여부
- str.contains("문자열")
- T, F로 출력된다.

```python
# 이메일에서 '.error'포함 여부 확인
condition = df['이메일'].str.contains('.error')

# 데이터 필터링
df.loc[condition]
```

### 문자열을 특정 문자열로 대체
- str.replace('기존 문자열', '대체 문자열')

```python
# 이메일의 '.error'를 '.com'으로 변환
df['이메일'] = df["이메일"].str.replace('.error', '.com')
df
```

### 특정 문자열로 시작/끝
- str,startswith('문자열') / str.endswith('문자열')

```python
# 성별에서 '남'으로 시작하는 데이터 확인
man = df['성별'].str.startswith('남')

# 데이터 필터링
df.loc[man]
```

### 구분자
- 특정 문자열을 기준으로 문자열을 분할
- str.split('특정 문자열')

```python
# 이메일을  '@'기준으로 분할
df['이메일'].str.split('@')

# 이메일을 '@' 기준으로 분할 후 첫번째 값 추출
df['아이디'] = df['이메일'].str.split('@').str[0]
df

# 두번째 값 추출
df['도메인'] = df['이메일'].str.split('@').str[1]
df
```

### 공백 제거
- str.strip() : 양쪽 공백 제거
- str.lstrip() : 왼쪽 공백 제거
- str.rstrip() : 오른쪽 공백 제거

```python
# 이름에서 양쪽 공백제거
df['이름'] = df['이름'].str.strip()
df
```

### 문자열의 길이계산
- str.len()
- 문자열의 길이 반환

```python
# Names 컬럼의 각 이름의 길이 계산
df['이름'].str.len()
```
